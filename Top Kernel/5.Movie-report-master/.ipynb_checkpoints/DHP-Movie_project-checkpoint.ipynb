# <a id="dtclean">2. Data cleaning </a>

```python
# Column: runtime
# Solve: replace NaN by mean = 106
df.loc[df.runtime.isnull(), 'runtime'] = 106

# Column: homepage
# Solve: replace NaN by an empty String
# df.loc[df.homepage.isnull(), 'homepage'] = ''
df.homepage.fillna('', inplace=True)
    
# Column: overview
# Solve: replace NaN by an empty String
df.loc[df.overview.isnull(), 'overview'] = ''
    
# Column: release_date
# Solve: Delete the row.
df.drop(df[df.release_date.isnull()].index, inplace=True)
# df.set_index(ind)

# Column: tagline
# Solve: replace NaN by an empty String
df.loc[df.tagline.isnull(), 'tagline'] = ''
```

```python
def json_to_col(dfr, col):
    for r in range(len(dfr)+1):
        try: 
            row = dfr[col].loc[r]
            series = pd.read_json(row).name
            for j in range(len(series)):
                dfr.loc[r, col+'_'+str(j)] = series[j]
        except AttributeError:
            continue
        except KeyError:
            continue
            
json_to_col(df, "production_countries")
# json_to_col(df, "genres")  # use One-hot-encode instead

# clean NaN data
for i in range(12):
        df.loc[df["production_countries_"+str(i)].isnull(), "production_countries_"+str(i)] = ''
# for i in range(7):
#         df.loc[df["genres_"+str(i)].isnull(), "genres_"+str(i)] = ''
```

# <a id="genres">4. The extent of effect that genres have on other categories</a>

```python
# To-do: replace this by pd.get_dummies
def json_to_one_hot_encode(dfr, col):
    for r in range(len(dfr)+1):
        try: 
            row = dfr[col].loc[r]
            series = pd.read_json(row).name
            for j in range(len(series)):
                cell = series[j]
                dfr.loc[r, cell] = 1
        except AttributeError:
            continue
        except KeyError:
            continue
            
json_to_one_hot_encode(df,"genres")

# clean NaN and convert column to int type
for i in df.columns[32:52]:
    df[i].fillna(0, inplace=True)
    df[i] = df[i].astype(np.int8)
```

```python
df.head(2)
```

```python
# create new df from genres chart then add 2 more columns
corr = df.iloc[:,32:52]
corr['revenue'] = df.revenue
corr['avg_score'] = df.vote_average
corr = corr.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            center=0)
```

From the graph above, the darker the joined cell's color, the lower the correlation between 2 columns. 
There is likely no correlation between revenue and genres, or average vote and genres. The brightest cell in revenue is from Adventure genres and the similar one in average vote is from Drama.

## Hypothesis testing:
For all the following tests, the significant level is:
$\alpha = 0.05$
### Relationship between Adventure genre & Revenue
$ H_0$ = The Adventure genre makes revenue increase <br>
$H_a$ = There is no relationship between those two

```python
stats.pearsonr(df.Adventure, df.revenue)
```

Adventure genre and Revenue correlation is below average and $p-value < \alpha$ <br>
=> It's statistically significant different between groups. Reject $H_0$ and accept $H_a$.
Since Adventure is the most likely to have highest affect to revenue. We can conclude that Genres do not affect revenue.

### Relationship between Drama genre & Vote Average
$ H_0$ = The Drama genre makes average score increase <br>
$H_a$ = There is no relationship between those two

```python
stats.pearsonr(df.Drama, df.vote_average)
```

Drama genre and Vote_average correlation is below average and $p-value < \alpha$ <br>
=> It's statistically significant different between groups. Reject $H_0$ and accept $H_a$.
Since Drama is the most likely to have highest affect to average score. We can conclude that Genres do not affect vote average.

# <a id="date">5. How release date influents on revenue</a>

```python
# convert to datetime format
df.release_date = pd.to_datetime(df.release_date)

# create a new weekday column
df['weekdays'] = df.release_date.dt.weekday

# convert weekdays in number to text:
weekday_names = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_dict = {key: weekday_names[key] for key in range(7)}
df['weekdays'] = df.weekdays.apply(lambda x: weekday_dict[x])

# convert y-m-d to m-d:
def date_to_day(date):
    y, m, d = str(date)[:10].split("-")
    return m + "-" + d
df['day'] = df.release_date.apply(date_to_day)

# convert y-m-d to m:
def date_to_month(date):
    y, m, d = str(date)[:10].split("-")
    return m
df['month'] = df.release_date.apply(date_to_month)
```

```python
weekday_sum = df.groupby("weekdays").sum()
weekday_sum = weekday_sum.loc[weekday_names]
weekday_sum['weekday'] = weekday_sum.index
# plot
ax = sns.barplot(x="weekday", y = "revenue", data=weekday_sum)
```

```python
day_sum = df.groupby("day").sum()
day_sum['day'] = day_sum.index
# day_sum
# plot
ax = sns.barplot(x="day", y = "revenue", data=day_sum)

# To-do: make this chart more clear
```

```python
month_sum = df.groupby("month").sum()
month_sum['month'] = month_sum.index
# day_sum
# plot
ax = sns.barplot(x="month", y = "revenue", data=month_sum)
```

From the charts, revenue will be the highest on May, June of each year and on Wednesday of each week.
